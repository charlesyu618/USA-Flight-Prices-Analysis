```python
import pandas as pd

data=pd.read_csv('processed_data.csv')
data = pd.DataFrame(data)
data
```

```python
print(data.shape)
percent_missing = (data.isnull().sum() / len(data)) * 100
print(f'percentage missing:{percent_missing}')
data.describe()
#considering we have 3 million samples, and our null values counts 7.4%, we will drop
```

```python
data = data.dropna()
data['searchDate'] = pd.to_datetime(data['searchDate'])
data['flightDate'] = pd.to_datetime(data['flightDate'])
data['daysBetween'] = (data['flightDate'] - data['searchDate']).dt.days
data=data.drop(['searchDate','legId','segmentsCabinCode','fareBasisCode',  'segmentsDepartureTimeEpochSeconds',
                'segmentsDepartureTimeRaw','segmentsArrivalTimeEpochSeconds','segmentsArrivalTimeRaw','segmentsArrivalAirportCode',
               'segmentsDepartureAirportCode','segmentsAirlineCode','segmentsDurationInSeconds','segmentsDistance'], axis=1)
data['month'] = data['flightDate'].dt.month
data['day_of_week'] = data['flightDate'].dt.dayofweek
data
```

```python
import re
def convert_duration_to_minutes(duration):
    hours = int(re.search(r'(\d+)H', duration).group(1)) if 'H' in duration else 0
    minutes = int(re.search(r'(\d+)M', duration).group(1)) if 'M' in duration else 0
    return hours * 60 + minutes

data['travelDurationMinutes'] = data['travelDuration'].apply(convert_duration_to_minutes)
data['#connect'] = data['segmentsAirlineName'].str.count(r'\|\|')
data=data.drop(['travelDuration'], axis=1)
data.replace({True: 1, False: 0}, inplace=True)
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(14,10))
data_heat = data.drop(columns = ['flightDate','startingAirport','destinationAirport','segmentsAirlineName','segmentsEquipmentDescription'])
cov_matrix = data_heat.cov()
sns.heatmap(cov_matrix, annot = True,fmt = '.2f',cmap = 'coolwarm')
plt.show()
```

```python
data = pd.DataFrame(data)
def keep_before_pipe(text):
    if isinstance(text, str):
        return text.split('||')[0]
    else:
        return text
data['segmentsAirlineName'] = data['segmentsAirlineName'].apply(keep_before_pipe)
#data = data.drop(columns = ['segmentsEquipmentDescription'])
data['segmentsEquipmentDescription'] = data['segmentsEquipmentDescription'].apply(keep_before_pipe)
```
