```python
import pandas as pd

data=pd.read_csv('processed_data.csv')
data = pd.DataFrame(data)
data['searchDate'] = pd.to_datetime(data['searchDate'])
data['flightDate'] = pd.to_datetime(data['flightDate'])
data['daysBetween'] = (data['flightDate'] - data['searchDate']).dt.days
data=data.drop(['searchDate', 'flightDate','legId','segmentsCabinCode','fareBasisCode',  'segmentsDepartureTimeEpochSeconds',
                'segmentsDepartureTimeRaw','segmentsArrivalTimeEpochSeconds','segmentsArrivalTimeRaw','segmentsArrivalAirportCode',
               'segmentsDepartureAirportCode','segmentsAirlineCode','segmentsDurationInSeconds','segmentsDistance'], axis=1)
```

```python
import re
def convert_duration_to_minutes(duration):
    hours = int(re.search(r'(\d+)H', duration).group(1)) if 'H' in duration else 0
    minutes = int(re.search(r'(\d+)M', duration).group(1)) if 'M' in duration else 0
    return hours * 60 + minutes

data['travelDurationMinutes'] = data['travelDuration'].apply(convert_duration_to_minutes)
data=data.drop(['travelDuration'], axis=1)
```

```python
data = pd.DataFrame(data)
def keep_before_pipe(text):
    if isinstance(text, str):
        return text.split('||')[0]
    else:
        return text
data['segmentsAirlineName'] = data['segmentsAirlineName'].apply(keep_before_pipe)
data['segmentsEquipmentDescription'] = data['segmentsEquipmentDescription'].apply(keep_before_pipe)
print(data)
```

```python
data=pd.read_csv('newdata.csv')
missing_values = data.isnull().sum()
duplicates = data.duplicated().sum()

(missing_values, duplicates)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded_segments = encoder.fit_transform(data['segmentsAirlineName'].values.reshape(-1, 1))
encoded_feature_names = encoder.get_feature_names_out(['segmentsAirlineName'])
encoded_segments_df = pd.DataFrame(encoded_segments, columns=encoded_feature_names)
data_without_segments = data.drop('segmentsAirlineName', axis=1)
data = pd.concat([data_without_segments, encoded_segments_df], axis=1)
```
